```python
def give_max(column, year):
    max_index = df[df['yr_built'] == year][column].nlargest(1).index[0]
    id = df.loc[max_index].id
    return id

for year in range(2000, 2016):
    print("The most expensive house in the year " + str(year) + " was house ID " + str(give_max('price', year)) + ".")
```

```python
for year in range(2000, 2016):
    print("The house with the most bathrooms in the year " + str(year) + " was house ID " + str(give_max('bathrooms', year)) + ".")
```

# Location analysis

```python
import math
```

```python
def calc_distance(loc1, loc2):
    return math.acos(math.sin(loc1[0] * math.pi/180) * math.sin(loc2[0] * math.pi/180) + math.cos(loc1[0] * math.pi/180) * math.cos(loc2[0] * math.pi/180) * math.cos(loc2[1] * math.pi/180 - loc1[1] * math.pi/180))  * 6371
```

```python
df.columns.get_loc('lat')
```

```python
df.columns.get_loc('long')
```

```python
df.iloc[0:, [0, 17, 18]]
```

```python
def distance_between(id1, id2):
    if df[df.id == id1]['id'].count() == 0 or df[df.id == id2]['id'].count() == 0:
        return "One or more IDs is non-existent"
    house1 = [list(df['lat'][df['id'] == id1])[0], list(df['long'][df['id'] == id1])[0]]
    house2 = [list(df['lat'][df['id'] == id2])[0], list(df['long'][df['id'] == id2])[0]]
    return (calc_distance(house1, house2))
```

```python
distance_between(7237550310, 3793500160)
```

```python
distance_between(1873100390, 8562750320)
```

```python
distance_between(1873100390, 8562750321)
```
